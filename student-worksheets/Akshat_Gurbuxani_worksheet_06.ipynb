### Kmeans ++

a) What is the difference between K means and K means ++?

K means initializes random centers. K means ++ is a more sophisticated method that combines randomness and FFT. It starts with a random center and selects the next center with probability proportional to D(x)^a where D(x) is the distance between x and the centers selected so far.

b) What are some limitations of K means ++?

Size    - It tends to split up in big clusters if they're uneven. Tends to favour clusters that have similar sizes.


Density - It tends to clump the clusters together rather than seperating them evenly.


Shape   - It tends to do well on globular shaped clusters. K means ++ performs inadequately on non-globular clusters.
